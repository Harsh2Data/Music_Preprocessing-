From 0.0 to 0.2, density is high:

```python
sns.boxplot(df['acousticness'])
```

```python
df[['track_name','artist_name','acousticness','popularity']][(df['acousticness'] >= 0.0) & (df['acousticness'] <= 0.2) ].sort_values(by='popularity',ascending=False)
```

Acousticness:
I analyzed songs across three acousticness ranges—low (0.0–0.2), medium (0.3–0.7), and high (0.8–1.0).

Low acousticness songs are mostly electronic or orchestral with little to no vocals or natural instruments.

Medium ones are a mix of vocals and instruments—balanced.

High acousticness songs are usually soft, unplugged, or instrumental with real instruments and no digital production.

I validated this by listening to songs from each range and confirming the sound style. This gave me a clear understanding of how this feature behaves.
